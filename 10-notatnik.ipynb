#### Indeksowanie stringów

* Możliwość **indeksowania** stringów, przy czym pierwszy znak to indeks `0`
* Brak oddzielnego typu znaku; znak to po prostu string o rozmiarze jeden:

````python
word = 'Python'
````

````python
word[0]  # znak na pozycji 0
````

````python
word[5]  # znak na pozycji 5
````

#### Ujemne indeksy

* Indeksy mogą być również liczbami ujemnymi, aby rozpocząć liczenie od prawej strony

````python
word[-1]  # ostatni znak
````

````python
word[-2]  # przedostatni znak
````

````python
word[-6]
````

* Zauważ, że ponieważ `-0` to to samo co `0`, ujemne indeksy zaczynają się od `-1`

#### Wycinanie (krojenie, plasterkowanie, kawałkowanie)

* Oprócz indeksowania obsługiwane jest również wycinanie
* Podczas gdy indeksowanie służy do uzyskiwania pojedynczych znaków, krojenie (wycinanie) pozwala uzyskać substring:

````python
word[0:2]  # znaki od pozycji 0 (dołączone) do 2 (wyłączone)
````

````python
word[2:5]  # znaki od pozycji 2 (dołączone) do 5 (wyłączone)
````

#### Przydatne wartości domyślne

* Indeksy plasterków mają przydatne wartości domyślne:
  * Pominięty pierwszy indeks przyjmuje wartość domyślną zero
  * Pominięty drugi indeks przyjmuje domyślnie rozmiar ciętego stringu

````python
word[:2]   # znak od początku do pozycji 2 (wykluczony)
````

````python
word[4:]   # znaki od pozycji 4 (dołączone) do końca
````

````python
word[-2:]  # znaki od przedostatniego (dołączone) do końca
````

* Zwróć uwagę, że początek jest zawsze uwzględniany, a koniec zawsze wykluczony
* Daje to pewność, że `s[:i] + s[i:]` jest zawsze równe `s`:

````python
word[:2] + word[2:]
````

````python
word[:4] + word[4:]
````

#### Sposób na zapamiętanie, jak działają plasterki

* Jednym ze sposobów na zapamiętanie, jak działają plasterki, jest myślenie o indeksach jako wskazujących **_między_** znakami, z lewą krawędzią pierwszego znaku numerowaną `0`
* Następnie prawa krawędź ostatniego znaku ciągu `n` znaków ma indeks `n`, na przykład:

```
 +---+---+---+---+---+---+
 | P | y | t | h | o | n |
 +---+---+---+---+---+---+
 0   1   2   3   4   5   6
-6  -5  -4  -3  -2  -1
```  
* Pierwszy rząd liczb podaje pozycje indeksów `0…6` w ciągu
* Drugi wiersz zawiera odpowiednie ujemne wskaźniki
* Kawałek od `i` do `j` składa się ze wszystkich znaków między krawędziami oznaczonymi odpowiednio `i` i `j`

#### Długość wycinka

* W przypadku indeksów nieujemnych długość wycinka jest różnicą indeksów, jeśli oba mieszczą się w granicach
* Na przykład długość `word[1:3]` wynosi `2`

````python
word[1:3]
````

#### Próba użycia zbyt dużego indeksu

* Próba użycia zbyt dużego indeksu spowoduje błąd:

## Biblioteki

Dodatkowe funkcje Pythona zawarte są w modułach, z których:
    
* Część jest dostarczona w każdej instalacji Pythona – tak zwana biblioteka „standardowa”;
* Reszta możliwa do doinstalowania – tak zwana biblioteka „zewnętrzna” (a raczej biblioteki „zewnętrzne”).


* Co to jest moduł?
  * W Pythonie dostępnych jest wiele modułów (czasami nazywanych bibliotekami)
  * Korzystając z tych modułów, możesz kodować znacznie szybciej
  * Pomyśl o nich jak o klockach konstrukcyjnych, zawierają duże zestawy funkcji (czasem klas), które zapewniają dodatkowe funkcje

* Moduły pomagają w uporządkowaniu kodu
* Zamiast jednego długiego pliku Pythona możesz mieć kilka plików (modułów)
* Moduł to plik Pythona zawierający (jedną lub więcej) funkcje lub klasy (o tym później)
* Program w języku Python może używać jednego lub więcej modułów

### Biblioteka „standardowa”

* Python posiada rozbudowaną **bibliotekę standardową**, umożliwiającą jego stosowanie do wielu zadań
* Twórcy języka stosują politykę tzw. **_Batteries Included_**, czyli dostarczenia wraz z pakietem instalacyjnym możliwie dużej liczby narzędzi
* Moduły standardowej biblioteki można uzupełniać modułami pisanymi:
  * C, lub
  * Pythonie

* Biblioteka standardowa jest szczególnie dobrze dostosowana do tworzenia aplikacji sieciowych, jako że obsługuje znaczną liczbę standardowych formatów i protokołów (np. **MIME**, **HTTP**)
* Dołączone są także moduły do tworzenia **GUI** (na bazie **Tcl**/**Tk**), obróbki **wyrażeń regularnych**, nawet prosty serwer **WWW** z obsługą **CGI**

* Większa część biblioteki standardowej dostępna jest na wszystkich platformach, dzięki czemu nawet duże aplikacje mogą często być uruchamiane bez konieczności modyfikacji:
  * Na **Uniksach**,
  * Pod **Windows**,
  * Na **Macintoshu**, i
  * Innych platformach
* Przeciwnie, niż np. dla **Javy**, nie ogranicza się zestawu dostępnych funkcji do części wspólnej dla różnych platform; np. na **uniksach** dostępna jest funkcja `os.fork()`, choć nie ma jej np. w **Windows**

### Biblioteki „zewnętrzne”

* Podobnie, jak w wypadku innych języków, opracowany został szereg standardów tworzenia **API** pomocniczego, np. sterowników **relacyjnych baz danych**
* Ze względu na w pełni dynamiczny system typów nie ma konieczności dołączania do biblioteki standardowej „interfejsu bazowego”, jak to ma miejsce np. w przypadku **JDBC**
* Twórca biblioteki zewnętrznej musi po prostu zapewnić, by stworzone przez niego moduły, funkcje i klasy posiadały odpowiednie atrybuty

````python
# lista
a = [ 5, 3, 2, -4, 12]

a.reverse()    # odwracanie
print(a)

a.sort()       # sortowanie
print(a)

a.append(3)    # dodawanie elementu na końcu
print(a)

del a[0]      # usuwanie elementu
print(a)
````

````python
# indeksowanie list
a = [ 5, 3, 1,6 ,3]

print(a[0])             # pierwszy element
print(a[-1])            # ostatni element
print(a[1:3])           # element od 1 do 2  (bez elementu 3!)
print(a[2:])            # element od 2 do końca
````

````python
# lista zawierająca listy
a= [[ 1, 2 ,3 ], [4, 5, 6, 7 ]]

print(len(a))
print(a[0])      # pierwszy wiersz
print(a[1][2])
````

### Słownik {klucz: wartość}

````python
# słownik { 'klucz1' : wartość1, 'klucz2': wartość2, ...}
s = { 'Ala' : 'ma kota', 'Ewa' : 'ma psa'}

print(s['Ala'])

s['Janek']='lubi Pythona'    # nowy element słownika

for k in s:
    print(k, s[k])

print(s.keys())      # lista kluczy
print(s.values())    # lista wartości
print(len(s))        # ilośc elementów

print('Ala' in s)    # czy klucz 'ala' jest w słowniku?
print('Zosia' in s)  # czy 'zosia' jest w słowniku?
````

### Serie w Pandas

*Seria* w bibliotece Pandas to jednowymiarowa tablica z indeksami, zawierająca dane tego samego typu (liczby całkowite, ciągi znaków, liczby zmiennoprzecinkowe, obiekty języka Python itp.). Jest to uogólniona tablica numpy z wyraźnie zdefiniowaną osią zwaną *indeksem*.

````python
# Przykład tworzenia serii w Pandas:
# Uporządkuj wszystkie elementy S1 razem
s1 = pd.Series([-3, -1, 1, 3, 5])
print(s1)
````

```
x = [10,20,30]
y = x[1]  # y --> 20
```
a także przy użyciu słownika w języku Python, w którym klucze pełnią rolę indeksu:
```
x = {'a':10, 'b':20, 'c':30}
y = x['b']  # y --> 20
```

### Dodaj kolumnę „Pora dnia”

Dodajmy kolumnę, która określa porę dnia (rano, po południu, wieczorem), w której dokonano sprzedaży.

Można to zrobić za pomocą funkcji [cut](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.cut.html) lub pisząc własną funkcję i używając [apply](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) do zastosowania tej funkcji do każdego wiersza kolumny Date_Time.

Oto jak działa funkcja `cut`, jeśli określamy przedziały i etykiety (istnieje wiele sposobów korzystania z tej funkcji — zobacz dokumentację):



`pd.cut(df[col], bins=[bin0,bin1,bin2,bin3], labels=['label0','label1','label2'])`

*bins* to zbiór elementów zawierający granice między etykietami. *labels* mogą być ciągami znaków, liczbami itp. – w zależności od potrzeb. Musi być o 1 etykietę więcej niż jest przedziałów. Dla 2 przedziałów i 3 etykiet:
* label0 jest używana dla `bin0 <= df[col] <= bin1`
* label1 jest dla `bin1 < df[col] <= bin2`
* label2 jest dla `bin2 < df[col] <= bin3`

Opcjonalny argument `right=True` zmienia znak `<` na `<=`, a `>=` na `>`. Jeśli nie znasz wartości minimalnej i maksymalnej, możesz użyć dla granic przedziałów wartości takich jak ujemna lub dodatnia nieskończoność:  `bins=[-np.inf,bin1,bin2,np.inf]`



````python
# Pandas.cut()
# 4 przedziały wartości, 3 etykiety.
# Kolumna typu Datetime posiada atrybut .dt, który umożliwia dostęp do funkcji i atrybutów związanych z datą i czasem
# Spróbuj: dir(df[»Date_Time«].dt)
df['Time_of_Day'] = pd.cut(df['Date_Time'].dt.hour,bins=[0,10, 16,24], labels=['morning','afternoon','evening'])
df
````

### Tablice Numpy

Każda kolumna ramki danych opiera się na tablicy Numpy. Przekazywanie danych z ramek danych do funkcji (np. z biblioteki scipy, własnych funkcji itp.) wymaga czasem przekształcenia kolumn ramki danych lub szeregów Pandas na ich reprezentację w postaci tablic Numpy. Istnieje wiele programów, które obsługują tablice Numpy jako dane wejściowe i nie obsługują biblioteki Pandas.

````python
def my_func(x):
    a = np.empty_like(x)
    b = 0
    for i in range(len(x)):
        b += x[i]
        a[i] = b
    return a

# Przekazywanie serii Pandas jest dopuszczalne. Często to działa, ale
# nie zawsze – na przykład jeśli funkcja „my_func” wywołuje skompilowany kod C++
# lub jest przyspieszana przez bibliotekę Numba, prawdopodobnie nie obsłuży
# tej serii poprawnie.
vals = my_func(flights['distance'])

# A może DataFrame? Usuń komentarz, a kod nie zadziała.
#vals2 = my_func(flights[['distance']])

# Wyślij DataFrame jako tablicę ndarray
vals3 = my_func(flights[['distance']].to_numpy())

# Konwersja dwóch kolumn na dwuwymiarową tablicę ndarray
numpy_vals = flights[['distance','air_time']].to_numpy()
print(numpy_vals.shape)
````
